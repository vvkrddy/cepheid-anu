```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# import mpld3
```

```python
wi = np.array([24, 16, 15, 19, 26, 28, 38, 35, 27, 20, 15, 19, 27, 32, 35, 31, 23, 18])
st = 38.0
# 57
area_wi = np.pi*(wi/2)**2
area_st = np.pi*(st/2)**2
mag = 5/2*np.log(area_st/area_wi)+5.7
jd = np.array([ 2449050,  2449110
,  2449150
,  2449180
,  2449240
,  2449300
,  2449350
,  2449375
,  2449435
,  2449500
, 2449540
,  2449635
,  2449700
, 2449740
,  2449760
,  2449800
,  2449870
,  2449950

 ])
```

```python
d = {'JD':jd, 'Mag':mag}
my_df = pd.DataFrame(data=d)
print(my_df)
print(max(my_df['Mag']) - min(my_df['Mag']))
print((my_df.iloc[10,0] - my_df.iloc[2,0] + my_df.iloc[14,0] - my_df.iloc[6,0])/2)
# print(my_df.iloc[14,0] - my_df.iloc[6,0])
```

```python
fig, ax = plt.subplots(1,1, constrained_layout=True)
# figsize=(15,10)

# y6 = np.array([1,1])

# fig.patch.set_facecolor('slategrey')
ax.set_axisbelow(True), 
ax.minorticks_on()
ax.grid(which='major', color="#838383", zorder=0, alpha=.4)
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black', zorder=0, alpha=0.2)
# ax.set_xlim([0,40])
# ax.set_ylim([f_transit-0.01, 1.01])
ax.set_ylabel('Magnitude', fontsize=20)
ax.set_xlabel('JD',fontsize=20)
# ax.set_xticks([2,4,6,8,10])

plt.setp(ax.get_xticklabels(), Fontsize=16)
plt.setp(ax.get_yticklabels(), Fontsize=16)

ax.plot(jd,mag,'-o')

# m, b = np.polyfit(x, y, 1)
# print(type(x))
# # plt.plot(x, y, 'o')
# plt.plot(x, x*m+b,c='red')

# fig.tight_layout()
plt.show()
```
