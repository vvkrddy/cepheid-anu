```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# import mpld3
```

```python
data1 = pd.read_excel('W UMa TESS data.xlsx', index_col=0)  
data2 = pd.read_excel('RRlyrae_TESS.xlsx', index_col=0)  

data1 = data1.dropna(subset=['LC_DETREND'])
data2 = data2.dropna(subset=['LC_DETREND'])
```

# RR Lyrae Phase Spectrum

```python
import numpy 
from matplotlib import pyplot  

fig, ax = pyplot.subplots(2,1, figsize=(20,5*4))
fig.patch.set_facecolor('white')


# assigning time values of the signal 
# initial time period, final time period and phase angle 
signalTime = data2['TIME'][2:1400]; 
  
# getting the amplitude of the signal 
signalAmplitude = data2['LC_INIT'][2:1400] 
  
# plotting the signal  
ax[0].plot(signalTime, signalAmplitude, color ='green') 
plt.setp(ax[1].get_xticklabels(), Fontsize=16)
plt.setp(ax[1].get_yticklabels(), Fontsize=16)

ax[0].set_xlabel('TIME(BJD)', fontsize=16)
ax[0].set_ylabel('LC_INIT',fontsize=16)
# plotting the phase spectrum of the signal  
ax[1].phase_spectrum(signalAmplitude, color ='green') 
ax[1].set_xlabel('FREQUENCY', fontsize=16)
ax[1].set_ylabel('PHASE radians',fontsize=16)
ax[0].set_title("Time vs LC_INIT", fontsize=16)
ax[1].set_title("Phase Spectrum", fontsize=16)
pyplot.suptitle("RR Lyrae", fontsize=16)


# pyplot.title("Phase Spectrum of the Signal") 
pyplot.show() 
```
